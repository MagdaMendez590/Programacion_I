## Ejercicio 1
Escriba una función que dibuje una rejilla como la siguiente:
![](rejilla.png)

```julia
function rejilla2x2()
    mas="+"		##### Defino los elementos mínimos de la rejilla
    menos="-"
    barra="|"
    espacio=" "
    for i in 1:11		#### Con este for recorro las líneas
        linea=""		#### Defino una cadena vacía donde voy a 
        for j in 1:11		#### Recorro las columnas
            if (i-1)%5 == 0
                if(j-1)%5 == 0	#### Condición de esquina exterior
                    linea*=mas
                elseif (j-1)%5 != 0  ### Condición de orilla horizontal
                    linea*=menos
                end
            else
                if(j-1)%5 == 0
                    linea*=barra     ### Condición orilla vertical
                else
                    linea*=espacio
                end
            end
        end
        println(linea)
    end

#### Uso la función

rejilla2x2()
```

```julia
#### Autor: Uriel ###

function adivinanza()
    num_secret = [ i for i in 1:100]     ## se define un arreglo en el cual estaran los numeros a adivinar
    random_num = rand(num_secret)       ## se define un número aleatorio del arreglo anterior
    numeros_comp = []                     ## Un array vacío donde se guardaran los numeros ingresados
while true                              ##Saludo                                               
        println("¿Quieres jugar?")                               
        comenzar = readline()
            if   comenzar == "si"
                println("Entonces adivina el número secreto")
            elseif comenzar == "no"
                println("Adios")
            break                       ## Condicion saludo que termina el codigo saludar y tambien termina el codigo en curso.
            end                         ## Fin del saludo.
    
                while true                                      ## Cuerpo del juego.
                    numero_adivinar = parse(Float64,readline())                  
                    if numero_adivinar < random_num           ## Evalua si el número dado es menor al numero a adivinar
                        println("el numero es más grande")
                         push!(numeros_comp,numero_adivinar) ## Guarda la colección de numeros ingresados en numeros_comp
               
           
                        
                        elseif numero_adivinar > random_num          ## Evalua si el numero dado es mayor al número a adivinar
                            println("el numero es más pequeño")
                            push!(numeros_comp,numero_adivinar)      ## Guarda la colección de numeros ingresados en numeros_comp
                
          
              
                        elseif numero_adivinar == random_num         ## Evalua si el numero dado es igual al número a adivinar
                            println(" Haz adivinado ")
                            break
                        end
                    end
                        
                     
                        intentos = length(Set(numeros_comp))    ## Set para no contar los numeros repetidos en el array numeros_comp
                        println("El número es $random_num")
                        println("Número de intentos fallidos :  $intentos ")      
                        println("¿Quieres volver?")
                        volver = readline() 
                        if  volver == "si"        ## regresar al saludo 
                            continue
                        elseif volver == "no "
                        end
                            println("Adios")
                        break
                
                end  
    
end
```

## Ejercicio  5

El matemático Srinivasa Ramanujan encontró una serie infinita que se puede usar para aproximar numéricamente el valor $\frac{1}{\pi}$: 
$$ \frac{1}{\pi} =  \frac{2\sqrt{2}}{9801}\sum_{k=0}^\infty \frac{(4k)!(1103+26390k)}{(k!)^4 396^{4k}}$$

1. Escriba una función que use esta expresión para calcular un estimado del valor de $\pi$. Deberá usar un loop `while` para calcular esta suma hasta que el valor del último término calculado sea menor a $10^{-15}$.

2. Haga un análisis gráfico de la convergencia de esta serie. Grafique el número de términos sumados y la discrepancia con el valor real de π.



```julia
function aprox_pi()
    i=0
    suma=BigFloat(0,512)
    terminoI=1
    aproximaciones = BigFloat[]   ### Hacemos un arreglo para guardar cada paso
    while abs(terminoI) >= 1e-14 && i<10
        terminoI = ((2^(3/2))/(9801))*(((factorial(BigFloat(4*i,512)))*(1103+26390*i))/((factorial(BigFloat(i,512))^4)*(396^(4*i))))
        suma += terminoI
        push!(aproximaciones,1/suma)
        i += 1
        println(terminoI)
        println("π=",1/suma)
    end
    #return suma
    return aproximaciones         ###Devolvemos un arreglo para poder graficar
end
```

```julia
aproximaciones_pi=aprox_pi()
```
